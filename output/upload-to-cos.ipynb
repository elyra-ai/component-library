```python
aws_access_key_id = os.environ.get('aws_access_key_id','7f1a60053c264c93b2f803744315ca29')
aws_secret_access_key = os.environ.get('aws_secret_access_key','cd2a15d12df5fb47e15e9e0ef06950b893ccdacbe551f573')
endpoint = os.environ.get('endpoint','https://s3.eu-de.cloud-object-storage.appdomain.cloud')
bucket_name = os.environ.get('bucket_name','rkie-deleteme-cos-standard-4u4')
source_file = os.environ.get('source_file','data.parquet')
destination_file = os.environ.get('destination_file','data.parquet')
data_dir = os.environ.get('data_dir', './data/')
```

```python
# override parameters received from a potential call using %run magic 
parameters = list(
    map(lambda s: re.sub('$', '"', s),
        map(lambda s: s.replace('=','="'),
            filter(
                lambda s: s.find('=') > -1,
                sys.argv
            )
        )
    )
)

for parameter in parameters:
    exec(parameter)
```

```python
client = ibm_boto3.client(
    's3',
    aws_access_key_id,
    aws_secret_access_key,
    endpoint_url=endpoint
)

client.upload_file(data_dir + source_file, bucket_name, destination_file)
```
