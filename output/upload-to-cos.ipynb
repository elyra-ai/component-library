```python
import os
import ibm_boto3
import sys
import re
import shutil
```

```python
aws_access_key_id = os.environ.get('aws_access_key_id')
aws_secret_access_key = os.environ.get('aws_secret_access_key')
endpoint = os.environ.get('endpoint')
bucket_name = os.environ.get('bucket_name')
source_file = os.environ.get('source_file')
destination_file = os.environ.get('destination_file')
data_dir = os.environ.get('data_dir', '../../data/')
```

```python
# override parameters received from a potential call using %run magic 
parameters = list(
    map(lambda s: re.sub('$', '"', s),
        map(lambda s: s.replace('=','="'),
            filter(
                lambda s: s.find('=') > -1,
                sys.argv
            )
        )
    )
)

for parameter in parameters:
    exec(parameter)
```

```python
client = ibm_boto3.client(
    's3',
    aws_access_key_id,
    aws_secret_access_key,
    endpoint_url=endpoint
)

client.upload_file(data_dir + source_file, bucket_name, destination_file)
```
